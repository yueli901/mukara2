```python
from evaluate_log import extract_metrics
import matplotlib.pyplot as plt

# Load and preprocess log data
df = extract_metrics('logs/training_log.log')[:]

# Define custom colors
train_color = (31/255, 119/255, 180/255)  # Blue for training
test_color = (214/255, 39/255, 40/255)    # Red for testing

# Define x-axis values
x_values = df['epoch'] + df['sensor'] / (max(df['sensor']) + 1)

# Create subplots in one row
fig, axes = plt.subplots(1, 1, figsize=(17, 4))

# Total Loss

# Traffic Loss (MGEH)
axes[0].plot(x_values, df['train_geh'], label='Train', color=train_color)
axes[0].plot(x_values, df['valid_geh'], label='Test', color=test_color)
axes[0].set_title('Traffic Loss (MGEH)')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('MGEH')
axes[0].legend()


# Adjust layout and save
plt.tight_layout()
plt.savefig('learning_curve.png')
plt.show()
```
