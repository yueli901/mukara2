```python
from evaluate_log import extract_metrics
import matplotlib.pyplot as plt

# Load and preprocess log data
df = extract_metrics('logs/training_log.log')[3:]

# Define custom colors
train_color = (31/255, 119/255, 180/255)  # Blue for training
test_color = (214/255, 39/255, 40/255)    # Red for testing

# Define x-axis values
x_values = df['epoch'] + df['sensor'] / (max(df['sensor']) + 1)

# Create subplots in one row
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

# Total Loss
axes[0].plot(x_values, df['train_total_loss'], label='Train', color=train_color)
axes[0].plot(x_values, df['valid_total_loss'], label='Test', color=test_color)
axes[0].set_title('Total Loss')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Loss')
axes[0].legend()

# Traffic Loss (MGEH)
axes[1].plot(x_values, df['train_geh'], label='Train', color=train_color)
axes[1].plot(x_values, df['valid_geh'], label='Test', color=test_color)
axes[1].set_title('Traffic Loss (MGEH)')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('MGEH')
axes[1].legend()

# Compactness Loss
axes[2].plot(x_values, df['train_compactness'], label='Train', color=train_color)
axes[2].plot(x_values, df['valid_compactness'], label='Test', color=test_color)
axes[2].set_title('Compactness Loss')
axes[2].set_xlabel('Epochs')
axes[2].set_ylabel('Loss')
axes[2].legend()

# Separation Loss
axes[3].plot(x_values, df['train_separation'], label='Train', color=train_color)
axes[3].plot(x_values, df['valid_separation'], label='Test', color=test_color)
axes[3].set_title('Separation Loss')
axes[3].set_xlabel('Epochs')
axes[3].set_ylabel('Loss')
axes[3].legend()

# Balance Loss
axes[4].plot(x_values, df['train_balance'], label='Train', color=train_color)
axes[4].plot(x_values, df['valid_balance'], label='Test', color=test_color)
axes[4].set_title('Balance Loss')
axes[4].set_xlabel('Epochs')
axes[4].set_ylabel('Loss')
axes[4].legend()

# Adjust layout and save
plt.tight_layout()
plt.savefig('learning_curve.png')
plt.show()
```
